### Importing and using modules

Let's start with importing and using what Python calls a **module**: A module is a file containing Python definitions and statements. For example, importing the **pandas** module gives us great tool to conduct data analysis.

```python
import pandas
```

### Inspecting data

Let's have a look at the data first by simply typing

```python
df
```

The rows in this table indicate *observations* in our data, the column names are *variables*. The "Readme.pdf" file in the replication package provides details on this data: It covers 3,722 county-district cells and contains amongst others the following variables:
- *d2_shnr_2002_Y* are the main outcome variables measuring the change in the Republican two-party vote share percentage between year 2002 and year Y = 2004, ..., 2016
- *d_imp_usch_pd* is the main independent variable for local labor market exposure to import competition from China
- *d_imp_otch_lag_pd* is the shift-share instrumental variable using Chinese exports to other countries
- *l_shind_manuf_cbp* is a control variable for manufacturing employment share

### Showing first descriptive statistics

The following *.describe* method gives summary statistics for each variable in the data. Note that the *.T* method transposes the table so that variable names are written as rows.

```python
df.describe().T
```

### Producing a histogram and scatter plot

An important visualisation of a distribution of a variable is a histogram. Scatter plots show how two variables relate to each other. To plot graphs in Python, we first import the **matplotlib.pyplot** module.

```python
import matplotlib.pyplot as plt
```

Note that we import this module "as plt", meaning that any method we use from this module can be called by writing plt.[method] instead of the long name of the module.

As a sidenote: You might wonder what methods are available in a given module. The following code will give you a list of methods that can be used with the matplotlib.pyplot module. As you can see, there is a host of methods within this module!

```python
 print (dir(plt))
```

You can also use the *help()* function in Python to get a more detailed description of a module or a method:

```python
help(plt.hist)
```

Let's use the *.hist* method to plot a histogram of the change in Republican vote share between 2002 and 2016. Note that one can access this variable in the data frame by using the [] brackets and calling the name of the variable with quotation marks:

```python
plt.hist(df["d2_shnr_2002_2016"])
```

Let's draw a scatter plot for the change in Republican vote share 2002-2016 on the x-axis and import competition on the y-axis.

```python
help(plt.scatter)
```

```python
plt.scatter(df["d2_shnr_2002_2016"], df["d_imp_usch_pd"])
```

Let's add some labels, change the colouring, and save to plot to your PC. Note that we import the *seaborn* module for colours.

```python
import seaborn as sns

# Define plot space
fig, ax = plt.subplots(figsize=(10, 6))

# Define x and y axes
ax.scatter(df["d2_shnr_2002_2016"], df["d_imp_usch_pd"], c="grey")

# Set plot title and axes labels
ax.set(title = "Globalisation and Voting",
       xlabel = "Delta Republican Vote Share",
       ylabel = "Import \n Competition")

# Save as .png
plt.savefig('C:/Users/felix/Dropbox/HfP/Teaching/SoSe21/Populism_Course/results/plot_1.png')
```
